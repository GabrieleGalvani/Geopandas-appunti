```
ax = quartieri.to_crs(epsg=3857).plot(edgecolor='k', facecolor='none')
bikes.to_crs(epsg=3857).plot(ax=ax,edgecolor='b', facecolor='none')
ctx.add_basemap(ax)
```

```
# calcolare la dimensione media dell'area dei quartieri di Milano
quartieri['Shape_Area'].mean()
```

```
# visualizzare i nomi dei quartieri piÃ¹ grandi della media
q_sup_media=quartieri[quartieri['Shape_Area']>quartieri['Shape_Area'].mean()]['NIL']
q_sup_media
```

```
# visualizzare i quartieri piÃ¹ grandi della media sulla mappa di Milano
```

```
ax = quartieri[quartieri['Shape_Area']>quartieri['Shape_Area'].mean()].to_crs(epsg=3857).plot(edgecolor='k', facecolor='none')
ctx.add_basemap(ax)
```

```
# visualizzare una mappa con la legenda in modo che i quartieri piÃ¹ grandi abbiano colori diversi rispetto ai quartieri piÃ¹ piccoli. 
# La scala dei colori deve essere rossa e si deve riuscire ad intravedere in trasparenza la mappa sottostante
```

```
ax = quartieri[quartieri['Shape_Area'] > quartieri['Shape_Area'].mean()].to_crs(epsg=3857).plot(column='Shape_Area',alpha=0.5,figsize=(12, 6),legend=True,edgecolor='k',cmap='OrRd')
ctx.add_basemap(ax)
```

```
# salvare in un geodataframe le informazioni sul quartiere di Greco, prendendole dal geodataframe dei quartieri
```

```
greco = quartieri.loc[quartieri['NIL'].str.contains('GRECO')]
greco
```

```
# trovare le piste ciclabili che passano per il quartiere di Greco
```

```
bikes[bikes.crosses(greco.geometry.squeeze())]
```

```
# visualizzare su una mappa il quartiere di Greco e le piste ciclabili che lo attraversano
# evidenziare bene in nero le piste ciclabili 
```

```
g1=greco.geometry.squeeze()
g2=bikes[bikes.intersects(g1)]
ax = greco.to_crs(epsg=3857).plot(figsize=(20,20),edgecolor='k',facecolor='none')
g2.to_crs(epsg=3857).plot(ax=ax,edgecolor='k',linewidth=5)
ctx.add_basemap(ax)
```
